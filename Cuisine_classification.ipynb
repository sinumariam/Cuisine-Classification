#### Text Cleaning to remove unnecessary characters and for lowercasing

* To remove unnecessary characters and for lowercasing

```python
# Define function for cleaning data
import re
def Pre_process_text(act_string):
    # remove all values other than alphabets
    new_string= re.sub(r'[^a-zA-Z0-9]'," ", act_string)
    # Lowercasing
    new_string = act_string.lower()
    return new_string
```

#### Apply TF-ID (Term Frequency-Inverse Document Frequency)

```python
# # Convert the restaurant details and cuisines into format that can be used for TF-ID extraction
data_frame['RestName_Cusines'] = data_frame['Restaurant Name'] + " " + data_frame['Cuisines']  .astype(str)
data_frame['RestName_Cusines'] =data_frame['RestName_Cusines'] .apply(str)
#Extract features from  text data using TF-IDF.
from sklearn.feature_extraction.text import TfidfVectorizer
X_Tfid= TfidfVectorizer(stop_words='english')
features = X_Tfid.fit_transform(data_frame['RestName_Cusines'])
```

```python
# Text Cleaning for column 'Cuisines'
data_frame['Cuisines']=data_frame['Cuisines'].apply(Pre_process_text)
```

```python
# Explore unique values in 'Cuisines'
target_names=data_frame['Cuisines'].unique()
target_names
```

```python
# No of unique values in Cuisines
data_frame['Cuisines'].nunique()
```

```python
# 'Cuisines' are Label Encoded
data_frame['Cuisine_Label'] = data_frame['Cuisines'].factorize()[0]
target = data_frame['Cuisine_Label']
```

#### Building the Model

```python
from sklearn.model_selection import train_test_split
```

* First step in building the model is to identify the Feature(Input) variables and Target (Output) variable
* Here features are the output obtained from TF_ID vectorization and Target is referred to 'Cuisine_Label'

#####  Splitting data for training and testing the model

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)
# Dimensions of Train and Test Data sets
print('Train set of features: ', X_train.shape)
print('Test set of features: ', X_test.shape)
print('Target for train: ', y_train.shape)
print('Target for test: ', y_test.shape)
```

###  1. Logistic Regression Model

```python
# Learn the Logistic Regression Model on train data
from sklearn.linear_model import LogisticRegression
my_model_LR=LogisticRegression()
```

```python
# Fitting the model in train data set ie the  Logistic Regression Model should learn from the Train Data
my_model_LR=my_model_LR.fit(X_train, y_train)
```

```python
# Predicting from Feature Test values
y_pred_LR = my_model_LR.predict(X_test)
y_pred_LR .shape
```

#### Testing the Logistic Regression Model

```python
# Print Classification Report
from sklearn.metrics import classification_report, accuracy_score
# Precision Score= TP/(TP+FP)
# Recall = TP/(TP+FN) ; Recall gives the percentage of positives well predicted by our model.
# f1_score= (Precition * Recall )/ (Precition + Recall)
print(classification_report(y_test, y_pred_LR)) 
```

```python
# find Accurancy Score = (TP+TN)/(TP+TN+FP+FN)
accuracy_score(y_test, y_pred_LR)
```

### 2. Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier
my_model_RF=RandomForestClassifier(random_state=10)
```

```python
# Fitting the Random Forest Classifier model in train data set
my_model_RF=my_model_RF.fit(X_train, y_train)
```

```python
# Predicting from Feature Test values
y_pred_RF = my_model_RF.predict(X_test)
```

#### Testing Random Forest Classifier Model

```python
# print Classification Report
# Precision Score= TP/(TP+FP)
# Recall = TP/(TP+FN) ; Recall gives the percentage of positives well predicted by our model.
# f1_score= (Precition * Recall )/ (Precition + Recall)
print(classification_report(y_test, y_pred_RF))
```

```python
# find Accurancy Score = (TP+TN)/(TP+TN+FP+FN)
accuracy_score(y_test, y_pred_RF)
```

* Random Forest has good Accuracy , Precision, Recall and f1 score compared to Logistic Regression
